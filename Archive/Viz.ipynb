```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os, glob
import regex as re
```

```python
# Declare the figure and axis.
fig, ax = plt.subplots(figsize=(16, 9))

# Read all the Starkey results and draw the boxplot.
labels = []
for i, filename in zip(range(0, 10), sorted(glob.glob('Starkey Results/*.txt'))):
    if 'baseline' not in filename:
        lab = re.sub('Starkey Results/starkeyResults', '', filename)
        labels.append(re.sub('.txt', '', lab))
        ax.boxplot(x=pd.read_csv(filename, header=None).iloc[:, 3], positions=[i],
                   vert=True, widths=0.60, patch_artist=True)
    else:
        lab = re.sub('Starkey Results/', '', filename)
        labels.append(re.sub('.txt', '', lab))
        ax.boxplot(x=pd.read_csv(filename, header=None), positions=[i],
                   vert=True, widths=0.60, patch_artist=True, boxprops=dict(facecolor='red'), notch=True)

ax.set_xticklabels(labels)
plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")
ax.set_title("Box Plots with comparison of different augmentation and point selection Combinations for Starkey Dataset")
# Display the chart.
fig.tight_layout()
```
